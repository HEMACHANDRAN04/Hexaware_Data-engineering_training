```python
courses_data = [
    (101, "Python Basics"),
    (102, "SQL Mastery"),
    (103, "Data Engineering"),
    (104, "Machine Learning"),
    (105, "Cloud Computing"),
    (106, "Power BI"),
    (107, "Java Programming"),
    (108, "Web Development"),
    (109, "DevOps"),
    (110, "PySpark")
]

courses_columns = [
    "course_id",
    "course_name"
]

courses_df = spark.createDataFrame(
    courses_data,
    courses_columns
)

courses_df.show()
```

```python
enrollments_data = [
    (1, 1, 101, "2025-02-01", "Active"),
    (2, 2, 102, "2025-02-02", "Completed"),
    (3, 3, 103, "2025-02-03", "Active"),
    (4, 4, 104, "2025-02-04", "Dropped"),
    (5, 5, 105, "2025-02-05", "Completed"),
    (6, 6, 106, "2025-02-06", "Active"),
    (7, 7, 107, "2025-02-07", "Completed"),
    (8, 8, 108, "2025-02-08", "Dropped"),
    (9, 9, 109, "2025-02-09", "Active"),
    (10, 10, 110, "2025-02-10", "Completed")
]

enrollments_columns = [
    "enrollment_id",
    "student_id",
    "course_id",
    "enrollment_date",
    "status"
]

enrollments_df = spark.createDataFrame(
    enrollments_data,
    enrollments_columns
)

enrollments_df.show()
```

```python
progress_data = [
    (1, 101, 45),
    (2, 102, 100),
    (3, 103, 65),
    (4, 104, 20),
    (5, 105, 100),
    (6, 106, 55),
    (7, 107, 100),
    (8, 108, 15),
    (9, 109, 40),
    (10, 110, 100)
]

progress_columns = [
    "student_id",
    "course_id",
    "completion_percentage"
]

progress_df = spark.createDataFrame(
    progress_data,
    progress_columns
)

progress_df.show()
```

```python
joined=students_df.join(enrollments_df,"student_id").join(courses_df,"course_id").join(progress_df,"student_id")
display(joined)
```

```python
final_df=joined.select("student_name",
                       "course_name",
                       "enrollment_date",
                       "completion_percentage"
                       )
final_df.show()
```

```python
final_df.createOrReplaceTempView(
    "final_view"
)
```

```python
%sql
SELECT * FROM final_view;
```

```python
final_df.write.format("delta").mode(
    "overwrite"
).saveAsTable("final_delta_table")
display(spark.table("final_delta_table"))
```
