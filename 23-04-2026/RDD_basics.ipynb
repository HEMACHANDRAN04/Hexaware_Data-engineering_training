```python
from pyspark.sql import SparkSession
spark=SparkSession.builder.appName("RDD").getOrCreate()
sc=spark.sparkContext
print("Spark is created")
```

```python
numbers=[10,20,30,40,50]
rdd=sc.parallelize(numbers)
print(rdd)
```

```python
rdd.collect()
```

```python
data=["arun","clara","thompson","jack","nora"]
data_rdd=sc.parallelize(data)
data_rdd.collect()
```

```python
rdd.getNumPartitions()
```

```python
rdd=sc.parallelize(numbers,3)
rdd.getNumPartitions()
```

```python
rdd=sc.parallelize([10,20,30,40,50,60],3)
rdd.glom().collect()
```

```python
rdd=sc.parallelize([1,2,3,4,5])
mapped_rdd=rdd.map(lambda x:x*2)
filtered_rdd=mapped_rdd.filter(lambda x:x>5)
print("Nothing executed")
```

```python
filtered_rdd.collect()
```

Lazy evaluation map filter only perform transformation when collect is called executed(it performs actions)
RDD operations: two categories
1. Transformations:
These create a new RDD.
Examples:
map
filter
flatMap
distinct
union
intersection
groupByKey
reduceByKey
2. Actions:
These return a result to the driver.
Examples:
collect
count
first
take
reduce

```python
rdd=sc.parallelize([1,2,3,4,5])
squared_rdd=rdd.map(lambda x:x*x)
squared_rdd.collect()
```

```python
rdd=sc.parallelize(["arun","clara","billy"])
upper_rdd=rdd.map(lambda x:x.upper())
upper_rdd.collect()
```

```python
rdd=sc.parallelize([10,15,20,25,30])
even_rdd=rdd.filter(lambda x:x%2==0)
even_rdd.collect()
```

```python
rdd=sc.parallelize(["arun","clara","william"])
length_rdd=rdd.filter(lambda x:len(x)>5)
length_rdd.collect()
```

```python
lines=sc.parallelize(["Python spark hadoop","spark sql pyspark"])
words=lines.flatMap(lambda line:line.split(" "))
words.collect()
```

```python
lines=sc.parallelize(["Hello world","Pyspark sql"])
result=lines.flatMap(lambda line:line.split(" "))
result.collect()
```

```python
lines=sc.parallelize(["Hello world","Pyspark sql"])
result=lines.map(lambda line:line.split(" "))
result.collect()
```

```python
rdd=sc.parallelize([10,20,30,10,20,30])
unique=rdd.distinct()
unique.collect()
```

```python
rdd=sc.parallelize([1,2,3])
rdd_2=sc.parallelize([4,5,6])
rdd.union(rdd_2).collect()
```

```python
rdd1=sc.parallelize([1,2,3,4,5])
rdd2=sc.parallelize([3,4,5,6,7])
rdd1.intersection(rdd2).collect()
```

```python
rdd1=sc.parallelize([1,2,3,4,5])
rdd2=sc.parallelize([4,5,6,7,8])
rdd1.subtract(rdd2).collect()
```

**reduceByKey and groupByKey**

```python
data=[
    ("IT",70000),
    ("HR",60000),
    ("IT",75000),
    ("Finance",80000),
    ("HR",58000)
]
rdd=sc.parallelize(data)
rdd.collect()
```

```python
rdd.reduceByKey(lambda x,y:x+y).collect()
```

```python
grouped=rdd.groupByKey()
[(k,list(v)) for k,v in grouped.collect()]
```

**mapValues and sortBy**

```python
rdd=sc.parallelize([("rahul",85),
                    ("zoe",95),
                    ("victoria",0)])
rdd.mapValues(lambda x:x+5).collect()
```

```python
rdd=sc.parallelize([50,80,10,20])
rdd.sortBy(lambda x:x).collect()
```

```python
lines=sc.parallelize(["python spark python",
                      "spark hadoop spark",
                      "python hadoop"])
words_count=(lines.flatMap(lambda line:line.split(" "))
.map(lambda word:(word,1))
.reduceByKey(lambda x,y:x+y)
)
words_count.collect()
```

```python
employees=sc.parallelize([
    ("Rahul",70000),
    ("Sneha",60000),
    ("Arjun",75000),
    ("priya",80000),
    ("Arun",50000)
])
high_salary=employees.filter(lambda x:x[1]>65000)
high_salary.collect()
```
