```python
from pyspark.sql import SparkSession
spark=SparkSession.builder.appName("RDD").getOrCreate()
sc=spark.sparkContext
print("Spark is created")
```

```python
logs = [
"Rahul login",
"Sneha login",
"Arjun login",
"Rahul purchase",
"Priya login",
"Sneha logout",
"Rahul login",
"Karan login",
"Arjun purchase",
"Sneha login",
"Rahul logout",
"Amit login",
"Priya purchase",
"Karan purchase",
"Sneha login",
"Rahul purchase",
"Meera login",
"Arjun login",
"Sneha purchase",
"Rahul login"
]

logs_rdd=sc.parallelize(logs)
logs_rdd.collect()
```

```python
sales = [
("Laptop",1,75000),
("Mouse",3,500),
("Keyboard",2,1500),
("Laptop",1,75000),
("Monitor",1,12000),
("Mouse",2,500),
("Keyboard",1,1500),
("Laptop",1,75000),
("Monitor",2,12000),
("Mouse",3,500),
("Laptop",1,75000),
("Keyboard",2,1500),
("Monitor",1,12000),
("Mouse",1,500),
("Laptop",1,75000)
]
sales_rdd=sc.parallelize(sales)
sales_rdd.collect()
```

```python
employees = [
("Rahul","IT",70000),
("Sneha","HR",60000),
("Arjun","IT",75000),
("Priya","Finance",80000),
("Karan","IT",50000),
("Amit","HR",58000),
("Meera","Finance",82000),
("Ravi","IT",72000),
("Neha","HR",61000),

("Vikram","Finance",90000)
]

employees_rdd=sc.parallelize(employees)
employees_rdd.collect()
```

```python
sentences = sc.parallelize([
"big data spark",
"spark hadoop hive",
"python spark hadoop",
"spark streaming kafka"
])
sentences.collect()
```

```python
words=sentences.flatMap(lambda word:word.split(" "))
words.collect()
```

```python
words.distinct().collect()
```

```python
count=(
    words.map(lambda x:(x,1))
    .reduceByKey(lambda x,y:x+y)
    )
count.collect()
```

```python
top_selling = sales_rdd.map(lambda x: (x[0], x[1])) \
                       .reduceByKey(lambda a,b: a+b) \
                       .sortBy(lambda x: x[1], ascending=False) \
                       .take(1)
print(top_selling)
```

```python
highest_revenue = sales_rdd.map(lambda x: (x[0], x[1]*x[2])) \
                         .reduceByKey(lambda a,b: a+b) \
                         .sortBy(lambda x: x[1], ascending=False) \
                         .take(1)
print(highest_revenue)
```

```python
more_than_3 = sales_rdd.map(lambda x: (x[0], x[1])) \
                       .reduceByKey(lambda a,b: a+b) \
                       .filter(lambda x: x[1] > 3) \
                       .collect()

print(more_than_3)
```
