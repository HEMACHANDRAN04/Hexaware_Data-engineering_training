Part 2 — DataFrame Transformations

```python
patients_df=patients_df.withColumn("registration_date", to_date(col("registration_date")))
```

```python
visits_df=visits_df.withColumn("visit_date", to_date(col("visit_date")))
```

```python
visit_doctor_df = visits_df.join(doctors_df,"doctor_id")

visit_doctor_df = visit_doctor_df.withColumn(
    "estimated_total_bill",
    col("consultation_fee") + col("test_cost")
)
visit_doctor_df.show()
```

```python
from pyspark.sql.functions import when, col
patients_df=patients_df.withColumn("age_category", when(col("age")<30, "young").when(col("age")<50, "adult").otherwise("senior"))


patients_df.show()
```

```python
patients_df = patients_df.withColumn(
    "membership_priority",
    when(col("membership")=="Gold",1)
    .when(col("membership")=="Silver",2)
    .otherwise(3)
)
patients_df.show()
```

```python
visits_df=visits_df.withColumn(
    "visit_priority",
    when(col("visit_status")=="Pending","high")
    .when(col("visit_status")=="Completed","medium")
    .otherwise("Low")
)
visits_df.show()
```

```python
visit_doctor_df=visit_doctor_df.withColumn(
    "high_bill_flag",
    when(col("estimated_total_bill")>3000,"yes")
    .otherwise("no")
)
visit_doctor_df.show()
```

```python
%sql
SELECT p.city,
SUM(pay.bill_amount) AS revenue
FROM patients p
JOIN visits v
ON p.patient_id=v.patient_id
JOIN payments pay
ON v.visit_id=pay.visit_id
GROUP BY p.city;
```

```python
%sql
SELECT d.specialization,
SUM(pay.bill_amount) AS revenue
FROM doctors d
JOIN visits v
ON d.doctor_id=v.doctor_id
JOIN payments pay
ON v.visit_id=pay.visit_id
GROUP BY d.specialization;
```

```python
%sql
SELECT * FROM payments
ORDER BY bill_amount DESC
LIMIT 5;
```

```python
%sql
SELECT doctor_id, COUNT(*) AS visit_count
FROM visits
GROUP BY doctor_id;
```

```python
%sql
SELECT patient_id, COUNT(*) AS visit_count
FROM patients
GROUP BY patient_id;
```

```python
%sql
    
SELECT payment_mode, AVG(bill_amount)
FROM payments
GROUP BY payment_mode;
```

```python
%sql
SELECT p.patient_name,
SUM(pay.bill_amount) AS total_bill
FROM patients p
JOIN visits v
ON p.patient_id=v.patient_id
JOIN payments pay
ON v.visit_id=pay.visit_id
GROUP BY p.patient_name
HAVING SUM(pay.bill_amount)>10000;
```

Part 6 — Window Functions

```python
patient_bill_df = final_df.groupBy("city","patient_name") \
                          .agg(sum("bill_amount").alias("total_bill"))

window_spec = Window.partitionBy("city").orderBy(col("total_bill").desc())

patient_bill_df.withColumn(
    "rank",
    rank().over(window_spec)
).show()
```

```python
doctor_bill_df=final_df.groupBy("specialization","doctor_name")\
    .agg(sum("bill_amount").alias("total_revenue"))
window_spec=Window.partitionBy("specialization").orderBy(col("total_revenue").desc())
doctor_bill_df.withColumn("rank",rank().over(window_spec)).show()
```

```python
patient_bill_df.withColumn(
    "row_num", row_number().over(window_spec)
).filter(col("row_num") == 1).show()
```

```python
w=Window.partitionBy("specialization").orderBy(col("consultation_fee").desc())
doctors_df.withColumn("rank",dense_rank().over(w)).show()
```

```python
doctor_bill_df.orderBy(col("total_revenue").desc()).show(2)
```

```python
daily_rev = final_df.groupBy("visit_date") \
                    .agg(sum("bill_amount").alias("daily_revenue"))

window_spec = Window.orderBy("visit_date")

daily_rev.withColumn(
    "running_total",
    sum("daily_revenue").over(window_spec)
).show()
```

```python
doctor_bill_df.withColumn(
    "running_total",
    sum("total_revenue").over(Window.orderBy("total_revenue"))
).show()
```

```python
city_rev = final_df.groupBy("city") \
                   .agg(sum("bill_amount").alias("revenue"))

city_rev.withColumn(
    "rank",
    rank().over(Window.orderBy(col("revenue").desc()))
).show()
```

```python
spec_rev = final_df.groupBy("specialization") \
                   .agg(sum("bill_amount").alias("revenue"))

spec_rev.withColumn(
    "rank",
    rank().over(Window.orderBy(col("revenue").desc()))
).show()
```

Compare Parquet and Delta
Parquet supports storage only
Delta supports ACID transactions
Delta supports UPDATE, DELETE, MERGE
Delta supports Time Travel and History
