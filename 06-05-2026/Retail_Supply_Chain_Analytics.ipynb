```python
daily_orders_data = [
(321,114,204,"2024-04-11",3,"Delivered"),
(322,118,210,"2024-04-11",2,"Delivered"),
(304,114,208,"2024-04-02",5,"Delivered"),
(319,112,208,"2024-04-10",2,"Delivered"),
(323,120,210,"2024-04-12",1,"Pending")
]
daily_orders_columns = [
"order_id",
"product_id",
"supplier_id",
"order_date",
"quantity",
"order_status"
]
daily_orders_df = spark.createDataFrame(daily_orders_data, daily_orders_columns)
```

```python
orders_df.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("orders_delta_target")
```

```python
initial_orders_df = spark.table("orders_delta_target")

initial_orders_df.show()
```

```python
daily_orders_df.createOrReplaceTempView("daily_orders_updates")
```

```python
from delta.tables import DeltaTable

target_table = DeltaTable.forName(spark, "orders_delta_target")

target_table.alias("target").merge(
    daily_orders_df.alias("source"),
    "target.order_id = source.order_id"
).whenMatchedUpdate(
    set = {
        "product_id": "source.product_id",
        "supplier_id": "source.supplier_id",
        "order_date": "source.order_date",
        "quantity": "source.quantity",
        "order_status": "source.order_status"
    }
).whenNotMatchedInsert(
    values = {
        "order_id": "source.order_id",
        "product_id": "source.product_id",
        "supplier_id": "source.supplier_id",
        "order_date": "source.order_date",
        "quantity": "source.quantity",
        "order_status": "source.order_status"
    }
).execute()
```

```python
spark.sql("""
SELECT * 
FROM orders_delta_target
WHERE order_id IN (304,319)
""").show()
```

```python
spark.sql("""
SELECT order_id, order_status
FROM orders_delta_target
WHERE order_id IN (304,319)
""").show()
```

```python
%sql
DESCRIBE HISTORY orders_delta_target
```

MERGE is important because:
It performs UPDATE + INSERT in a single operation.
Helps implement incremental loading.
Avoids duplicate records.
Efficient for handling daily transaction updates.
Supports ACID transactions in Delta Lake.
Commonly used in real-time ETL pipelines.

Compare Parquet vs Delta behavior
Parquet
Supports only storage
No ACID transactions
No update/delete support
Delta
Supports ACID transactions
Supports update/delete/merge
Supports time travel and histor
