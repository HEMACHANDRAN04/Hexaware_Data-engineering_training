Section 8 — Window Functions

```python
customers.write.mode("overwrite").saveAsTable("customers_details")
products.write.mode("overwrite").saveAsTable("products_details")
orders.write.mode("overwrite").saveAsTable("orders_details")
order_items.write.mode("overwrite").saveAsTable("order_items_details")
payments.write.mode("overwrite").saveAsTable("payments_details")
```

```python
spark.sql("""
SELECT
product_id,
product_name,
category,
price,
RANK() OVER (ORDER BY price DESC) AS price_rank
FROM products_details
""").show()
```

```python
spark.sql("""SELECT product_name, category, price, RANK() OVER(PARTITION BY category ORDER BY price DESC) AS price_rank
FROM products_details
""").show()
```

```python
spark.sql("""
SELECT customer_id,
name,
city,
age,
signup_date,
ROW_NUMBER() OVER (ORDER BY signup_date) AS row_num
FROM customers_details
""").show()
```

```python
spark.sql("""
SELECT
c.customer_id,
c.name,
SUM(oi.quantity * p.price) AS revenue,
RANK() OVER (ORDER BY SUM(oi.quantity * p.price) DESC) AS revenue_rank
FROM customers_details c
JOIN orders_details o ON c.customer_id = o.customer_id
JOIN order_items_details oi ON o.order_id = oi.order_id
JOIN products_details p ON oi.product_id = p.product_id
GROUP BY c.customer_id, c.name
""").show()
```

```python
spark.sql("""
SELECT
c.customer_id,
c.name,
SUM(oi.quantity * p.price) AS revenue,
DENSE_RANK() OVER (ORDER BY SUM(oi.quantity * p.price) DESC) AS dense_rank
FROM customers_details c
JOIN orders_details o ON c.customer_id = o.customer_id
JOIN order_items_details oi ON o.order_id = oi.order_id
JOIN products_details p ON oi.product_id = p.product_id
GROUP BY c.customer_id, c.name
""").show()
```

```python
spark.sql("""
SELECT
order_id,
payment_type,
amount,
RANK() OVER (ORDER BY amount DESC) AS payment_rank
FROM payments_details
""").show()
```

```python
spark.sql("""
SELECT c.name, SUM(oi.quantity * p.price) AS revenue
FROM customers c
JOIN orders o ON c.customer_id = o.customer_id
JOIN order_items oi ON o.order_id = oi.order_id
JOIN products p ON oi.product_id = p.product_id
GROUP BY c.name
ORDER BY revenue DESC
LIMIT 3
""").show()

```

```python
spark.sql("""
SELECT p.product_name, SUM(oi.quantity * p.price) AS revenue
FROM order_items oi JOIN products p
ON oi.product_id = p.product_id
GROUP BY p.product_name
ORDER BY revenue DESC
LIMIT 1
""").show()
```
