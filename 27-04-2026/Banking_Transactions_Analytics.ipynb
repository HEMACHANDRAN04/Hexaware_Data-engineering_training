Section 10 — Window Functions

```python
customers.createOrReplaceTempView("customers")
accounts.createOrReplaceTempView("accounts")
transactions.createOrReplaceTempView("transactions")
branches.createOrReplaceTempView("branches")
```

```python
spark.sql("""
SELECT c.name,c.city,a.balance,
RANK() OVER (PARTITION BY c.city ORDER BY a.balance DESC) AS rank
FROM customers c
JOIN accounts a ON c.customer_id=a.customer_id
""").show()
```

```python
spark.sql("""
SELECT * FROM (
    SELECT c.city, c.name, a.balance,
    ROW_NUMBER() OVER (PARTITION BY c.city ORDER BY a.balance DESC) AS rn
    FROM customers c
    JOIN accounts a ON c.customer_id = a.customer_id
) t
WHERE rn = 1
""").show()
```

```python
spark.sql("""SELECT txn_id, txn_type, amount,
DENSE_RANK() OVER(PARTITION BY txn_type ORDER BY amount DESC) AS dense_rank
FROM transactions
""").show()
```

```python
spark.sql("""
SELECT * FROM (
    SELECT account_id, txn_type, amount,
    RANK() OVER (PARTITION BY account_id ORDER BY amount DESC) AS rnk
    FROM transactions
) t
WHERE rnk <= 2
""").show()
```

```python
spark.sql("""
SELECT account_id, txn_date, amount,
SUM(amount) OVER (PARTITION BY account_id ORDER BY txn_date) AS running_total
FROM transactions
""").show()
```

```python
spark.sql("""
SELECT * FROM (
    SELECT c.city, c.name, MAX(t.amount) AS max_txn,
    RANK() OVER (PARTITION BY c.city ORDER BY MAX(t.amount) DESC) AS rnk
    FROM customers c
    JOIN accounts a ON c.customer_id = a.customer_id
    JOIN transactions t ON a.account_id = t.account_id
    GROUP BY c.city, c.name
) t
WHERE rnk = 1
""").show()
```

```python
spark.sql("""
SELECT * FROM (
    SELECT b.region, c.name, a.balance,
    RANK() OVER (PARTITION BY b.region ORDER BY a.balance DESC) AS rnk
    FROM customers c
    JOIN accounts a ON c.customer_id = a.customer_id
    JOIN branches b ON a.branch = b.branch
) t
WHERE rnk = 1
""").show()
```
