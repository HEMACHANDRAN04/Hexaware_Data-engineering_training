```sql
SELECT *
FROM employees e
WHERE salary > (
    SELECT AVG(salary)
    FROM employees
    WHERE department = e.department
);
```

```sql
SELECT *
FROM employees e
WHERE experience = (
    SELECT MAX(experience)
    FROM employees
    WHERE department = e.department
);
```

```sql
CREATE TABLE employee_salary_analysis AS
SELECT
    emp_name,
    city,
    department,
    salary,
    experience,
    CASE
        WHEN salary >= 95000 THEN 'High'
        WHEN salary >= 80000 THEN 'Medium'
        ELSE 'Low'
    END AS salary_category,
    AVG(salary) OVER (PARTITION BY department) AS department_avg_salary
FROM employees;
```
